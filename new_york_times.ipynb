```python
import requests
from bs4 import BeautifulSoup
session = requests.Session()
import datetime
import time

def NYDataCollect(searchword):
    url = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q="+searchword+"&api-key=6dqBKUiy4CGD7qvuIp6KoGfj4KUDzGvR"
    r = requests.get(url)
    data = r.json()
    # print(data)
    return data["response"]["docs"]

def urlCollector(inpData):
    urlList = []
    for i in inpData:
        urlList.append(i["web_url"])
    return urlList

def articleText(url):
    try:
        req = session.get(url)
        soup = BeautifulSoup(req.text,"lxml")
        paragraphs = soup.find_all('p', class_='story-body-text story-content')
        article = ""
        for i in paragraphs:
            article = article + i.get_text() 
            writeToFile(article,"nyData")
    except:
        print("url problem for " + url)

def writeToFile(inpStr,folder):
    filename = "./"+ folder+ "/article_" + datetime.datetime.fromtimestamp(time.time()).strftime('%Y_%m_%d_%H_%M_%S') + ".txt"
    dFile= open(filename,"w")
    dFile.write(inpStr)
    dFile.close()

def getNYDataByTopic(topic):
    topicdata = NYDataCollect(topic)
    urls = urlCollector(topicdata)
    for i in urls:
        print(i)
        articleText(i)

getNYDataByTopic("sports")
getNYDataByTopic("soccer")
getNYDataByTopic("football")
getNYDataByTopic("golf")
getNYDataByTopic("baseball")
getNYDataByTopic("basketball")
getNYDataByTopic("boxing")
```
